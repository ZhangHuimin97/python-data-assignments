```python
def get_rooms(soup):
    rooms = []
    for i in soup.find_all('div',attrs={'class':"_gig1e7"}):  
        room = {}
        
        style=i.find('small',attrs={'class':"_f7heglr"}).text.split()
        if len(style)<3:
            layout="None"
        else:
            layout=style[2]
        
        priceall=i.find('span', attrs={'class':"_xbdrhpb"}).text
        
        equipmentall=i.find('div',attrs={'class':"_2q1l5fu"}).text
        equipmentm=equipmentall.replace('等多项便利设施','')
        equipmentt=equipmentm.replace('·','')
        
        
        
        room['name'] =i.find('div', attrs={'class':"_4t0rf0r"}).text#.encode('utf8')#尝试解码
        room['feature'] =style[0]
        room['layout'] =layout
        
        room['price'] = priceall[3:6]
        room['comments'] = i.find('span', attrs={'class':"_1cy09umr"}).text
        room['equipment'] = equipmentt
        
        
        rooms.append(room)
        
    return rooms
```

```python
list1=[18,36,54,72,90,108]
all_rooms_combine=[]
for a in list1:
    url="https://zh.airbnb.com/s/Seoul--South-Korea/homes?refinement_paths%5B%5D=%2Fhomes&adults=1&allow_override%5B%5D=&children=0&guests=1&infants=0&place_id=ChIJzWXFYYuifDUR64Pq5LTtioU&query=Seoul%2C%20South%20Korea&s_tag=QU_HsI2L&section_offset=7&items_offset={0}".format(a)
    browser.get(url)
    time.sleep(2)
    html=browser.page_source
    soup = bs4.BeautifulSoup(html,'html.parser')
    all_rooms=get_rooms(soup)
    time.sleep(2)
    if all_rooms in all_rooms_combine:
        all_rooms_combine= all_rooms_combine
    else:
        all_rooms_combine= all_rooms_combine+all_rooms
    print(all_rooms_combine)
    
    
```

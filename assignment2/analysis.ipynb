# 1.What is the most common type of house in seoul on Airbnb?

## Apartment ranks the top.

```python
feature= df['Feature'].value_counts()
plt.title('The amount of feature')
plt.xticks(range(len(feature.index)),feature.index,fontproperties='SimHei',color='red')
plt.plot(range(len(feature.index)),feature.values)
```

## But which kind of layout is the most common?
## Let's find out.

```python
layout = df['Layout'].value_counts()[:10]
plt.title('The amount of layout')
plt.yticks(range(len(layout.index)),layout.index,fontproperties='SimHei',color='red')
plt.barh(range(len(layout)),layout.values,align='center')
```

## From the chart above, we can see that the layout with 1 room, 1 bathroom, 2 beds ranking the top.
## We can infer that single-room is the most popular.

# 3.What's the relationship between price and equipment ?

```python
df = pd.read_csv('result.csv',header=None,names=['Name','Feature','Layout','Price','Comment','Equipment'])
len(df['Equipment'])
```

```python
num = []
for eq in df['Equipment']:
    eq = str(eq).split('  ')
    i = 0
    for e in eq:
        if len(e) != 0:
            i = i + 1
    num.append(i)
df['Num of equipment'] = num
```

```python
df['Price'].corr(df['Num of equipment'])
```

## From the above, we find that the price is nothing to do with the variety of equipment, which is surprising.

# 5.Conclusion

## From our analysis, we find that most houses in Seoul on Airbnb are prepared for the couple or single person, and there is no relationship between the price and the variety of equipment.
